**Pregunta 1**

Considerando que la transpuesta de una matriz es el resultado de cambiar las filas a columnas. Sea una matriz $A\in\mathbb{R}^{m\times n}$, su transpuesta se escribe como $A^{T}\in\mathbb{R}^{n\times m}$ y sus entradas están dadas por:

\begin{equation}
\left(A^{T}\right)_{i,j}=A_{j,i}.
\end{equation}

Proponga Usted dos matrices, $A\in\mathbb{R}^{4\times 3}$ y $B\in\mathbb{R}^{3\times 4}$ (no las busque en Internet, trate de proponer los valores por si mismo), y utilizando numpy muestre que las siguientes propiedades se cumplen con sus matrices propuestas.

**Note** que ni A ni B son cuadradas, y no tienen las mismas dimensiones.


**1.1 (15 puntos)**

- $\left(A^{T}\right)^{T}=A$


```python
###############################################################################
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
###############################################################################

A = np.array ([[1, 2, 3],
              [2, 4, 6],
              [4, 8, 12],
              [8, 16, 24]
             ])

At = np.transpose(A)

Att = np.transpose(At)

print(f"Matriz A: \n{A}")
print("--------------")
print(f"Matriz transpuesta de A:\n {At}")
print("--------------")
print(f"Matriz transpuesta de transpuesta de A: \n {Att}")
print("--------------")
#print(f"(At)t == A => \n { np.equal(A, Att)}")
print(f"(Aᵗ)ᵗ == A => { np.array_equal(A, Att)}")
```

**1.2 (15 puntos)**

- $\left(AB\right)^{T}=B^{T}A^{T}$


```python
###############################################################################
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
###############################################################################

A = np.array ([[1, 2, 3],
              [2, 4, 6],
              [4, 8, 12],
              [8, 16, 24]
             ])

B = np.array ([[1, 2, 3, 4],
              [3, 6, 9, 12],
              [9, 18, 27, 36],
             ])

AB = np.matmul(A, B)
ABt = np.transpose(AB)

print(ABt)
print("--------------")

Bt = np.transpose(B)
At = np.transpose(A)

BtPlusAt = np.matmul(Bt, At)

print (BtPlusAt)
print("--------------")

print(f"(AB)ᵗ == (Bᵗ)(Aᵗ) => {np.array_equal(ABt, BtPlusAt)}")
```

**1.3 (15 puntos)**

Para este ejercicio, **proponga dos matrices de igual dimensionalidad**


- $\left(A+B\right)^{T}=A^{T}+B^{T}$.


```python
###############################################################################
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
###############################################################################

A = np.array ([[1, 2, 3],
              [2, 4, 6],
              [4, 8, 12],
             ])

B = np.array ([[1, 2, 3],
              [3, 6, 9],
              [9, 18, 27],
             ])

APlusB = np.add(A, B)
APlusB_t = np.transpose(APlusB)

print(APlusB_t)
print("--------------")

At = np.transpose(A)
Bt = np.transpose(B)

AtPlusBt = np.add(At, Bt)

print(AtPlusBt)
print("--------------")

print(f"(A+B)ᵗ == Aᵗ + Bᵗ => {np.array_equal(APlusB_t, AtPlusBt)}")
```

De lo contrario, si uno de los vectores en tal conjunto puede ser representado como una combinación lineal del resto de vectores, entonces los vectores son linealmente dependientes, lo que se expresa como:

\begin{equation}
\vec{x}_{j}=\sum_{i=1}^{n-1}\alpha_{i}\vec{x}_{i}
\end{equation}

para cualquier conjunto de valores escalares $\alpha_{1},\ldots,\alpha_{n-1}\in\mathbb{R}$ se dice que el vector $\vec{x}_{j}\in\mathbb{R}^{m}$ es linealmente dependiente de los vectores $\vec{x}_{i}$.
