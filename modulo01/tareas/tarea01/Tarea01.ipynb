a)  
$g\left(t\right)=\log _{10} u(t)$







R/ Para que el sistema sea lineal debe cumplir:

1. **Aditividad**
2. **Homogeneidad**

---

#### 1) Propiedad de Aditividad

Sean $u_1(t)$ y $u_2(t)$ dos entradas.

Evaluamos:

$$
L\{u_1(t) + u_2(t)\} = \log_{10}\left(u_1(t) + u_2(t)\right)
$$

Si fuera lineal, debería cumplirse:

$$
L\{u_1(t)\} + L\{u_2(t)\}
=
\log_{10}(u_1(t)) + \log_{10}(u_2(t))
$$

Pero sabemos que:

$$
\log_{10}(u_1(t)) + \log_{10}(u_2(t))
=
\log_{10}\left(u_1(t) u_2(t)\right)
$$

Y en general:

$$
\log_{10}\left(u_1(t) u_2(t)\right)
\neq
\log_{10}\left(u_1(t) + u_2(t)\right)
$$

Por lo tanto, **no se cumple la propiedad de aditividad**.

---

#### 2) Propiedad de Homogeneidad

Sea $\alpha$ una constante.

Evaluamos:

$$
L\{\alpha u(t)\}
=
\log_{10}(\alpha u(t))
$$

Si fuera lineal, debería cumplirse:

$$
\alpha L\{u(t)\}
=
\alpha \log_{10}(u(t))
$$

Pero:

$$
\log_{10}(\alpha u(t))
=
\log_{10}(\alpha) + \log_{10}(u(t))
\neq
\alpha \log_{10}(u(t))
$$

Por lo tanto, **no se cumple la propiedad de homogeneidad**.

---

### Conclusión

El sistema

$$
g(t) = \log_{10} u(t)
$$

**no es lineal**.

---

### a) Respuesta a)

Sea el vector:

$$
\mathbf{a} =
\begin{bmatrix}
-9 \\
7
\end{bmatrix}
$$

La norma $L_2$ se define como:

$$
\|\mathbf{a}\|_2
=
\sqrt{\sum_{i=1}^{n} a_i^2}
$$

Entonces:

$$
\|\mathbf{a}\|_2
=
\sqrt{(-9)^2 + 7^2}
$$

$$
=
\sqrt{81 + 49}
$$

$$
=
\sqrt{130}
$$

---

#

```python
from math import sqrt

def normal2(vector):
    suma = 0
    for i in vector:
        suma += i**2
    return sqrt(suma)

a = np.array([9, -7], dtype=np.int16)

print(f"Resultado programado mediante una funcion -> {normal2(a)}")

print(f"Resultado programado mediante la función norm -> {np.linalg.norm(a)}")

```

¿Existen atributos colineanes? ¿Cuáles? Programe en python para mostrar su respuesta.



```python


separador = "-"*40
archivo = "Datos_01.csv"

# carga el archivo en pandas
dataFrame = pd.read_csv(archivo, header = 0, delimiter=';')

# guarda el nombre de las columnas en una lista
colNames = dataFrame.columns


# muestra los primeros elementos del dataFrame
print(separador)
print("Datos en dataFrame:")
print(dataFrame.head() )

# Convertir de pandas a numpy
datos = pd.DataFrame(dataFrame).to_numpy().astype(np.float64)

# En cada vector columna hay un atributo
# El atributo1 está en datos[:,0]
# El atributo2 está en datos[:,1]
# y así sucesivamente

# Programar para determinar vectores colineales


def angle_between_vectors(u, v): # Función encontrada en stackoverflow, que calcula el ángulo entre dos vectores utilizando el producto punto y las normas de los vectores
    dot_product = np.dot(u, v)
    norm_u = np.linalg.norm(u)
    norm_v = np.linalg.norm(v)

    if norm_u == 0 or norm_v == 0:
        return 0.0

    cos_theta = dot_product / (norm_u * norm_v)

    # Corrección numérica
    cos_theta = np.clip(cos_theta, -1.0, 1.0)

    angle_rad = acos(cos_theta)
    angle_deg = degrees(angle_rad)

    return angle_deg



def verificar_colineales(*vectores): # Toma la cantidad de vectores que le pasen como argumentos, para hacer la función mas general
    vectores_colineales = []
    try:
        for i in range(len(vectores)):
            for j in range(i+1, len(vectores)):  # Esta iteracion anidada es para comparar cada vector con los demás sin repetir comparaciones ejemplo: comparar el vector 0 con el 1, luego el 0 con el 2, y así sucesivamente, pero no comparar el vector 1 con el 0 nuevamente
                if vectores[i].shape != vectores[j].shape:
                    raise ValueError(f"Los vectores {i} y {j} tienen diferentes dimensiones, no se pueden comparar.")
                angulo = angle_between_vectors(vectores[i], vectores[j])

                print(f"Ángulo entre {i} y {j}: {angulo}")

                if abs(angulo) < 1e-8 or abs(angulo - 180) < 1e-8:
                    vectores_colineales.append((i, j))

        return vectores_colineales

    except ValueError as e:
        print(e)
        return []
    



print(verificar_colineales(datos[:,0], datos[:,1], datos[:,2], datos[:,3], datos[:,4]))

datos_sin_colineales = np.delete(datos, 3, axis=1).astype(np.float64) # Se elimina la columna 3 del array de datos, que corresponde al vector que es colineal con otro, para obtener un nuevo array sin vectores colineales.

datos_nocoli_pandas = pd.DataFrame(datos_sin_colineales, columns=colNames.drop(colNames[3])) # Se actualizan los nombres de las columnas del nuevo DataFrame para reflejar la eliminación de la columna colineal.
print(f"Datos sin vectores colineales:\n{datos_nocoli_pandas}")
```

b)

En el archivo "reales.csv" se encuentran los valores reales de un conjunto de datos, y en el archivo "predicciones.csv" lo que predijo un algoritmo.

Cargue ambos archivos y muestre la norma 2 y la norma 5, de la diferencia entre el real y el predicho.

Finalmente, grafique el predicho y el real en un mismo gráfico para comparar.

```python

# -----------------------------
# Cargar archivos
# -----------------------------

archivo = "reales.csv"
dataFrameReales = pd.read_csv(archivo, header=0)

archivo = "predicciones.csv"
dataFramePredichos = pd.read_csv(archivo, header=0)

# -----------------------------
# Convertir de pandas a numpy
# -----------------------------

# Si los archivos tienen una sola columna:
reales = dataFrameReales.values.flatten()
predichos = dataFramePredichos.values.flatten()

# Verificar que tengan el mismo tamaño
if reales.shape != predichos.shape:
    raise ValueError("Los vectores no tienen la misma dimensión.")

# -----------------------------
# Calcular diferencia
# -----------------------------

diferencia = reales - predichos

# -----------------------------
# Calcular normas
# -----------------------------

norma_2 = np.linalg.norm(diferencia, ord=2)
norma_5 = np.linalg.norm(diferencia, ord=5)

print("Norma 2 de la diferencia:", norma_2)
print("Norma 5 de la diferencia:", norma_5)

# -----------------------------
# Graficar
# -----------------------------

plt.figure(figsize=(8,5))
plt.plot(reales, label="Reales")
plt.plot(predichos, label="Predicciones")
plt.title("Comparación: Reales vs Predicciones")
plt.xlabel("Índice")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```
